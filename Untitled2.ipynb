```python

# Install necessary libraries (uncomment if running for the first time)
# !pip install matplotlib seaborn

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample data (you can update these values based on real sources)
ai_data = {
    'AI Technology': [
        'ChatGPT (OpenAI)',
        'Google Gemini (Bard)',
        'Claude (Anthropic)',
        'Siri & Alexa',
        'Midjourney/DALL·E',
        'GitHub Copilot',
        'Tesla Autopilot'
    ],
    'Usage Rate (%)': [35, 20, 10, 8, 12, 9, 6]
}

# Convert to DataFrame
df = pd.DataFrame(ai_data)

# Set plot style
sns.set(style="whitegrid")

# Bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Usage Rate (%)', y='AI Technology', data=df, palette='coolwarm')
plt.title('Usage Rate of Different AI Technologies in 2024')
plt.xlabel('Usage Rate (%)')
plt.ylabel('AI Technology')
plt.tight_layout()
plt.show()

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(df['Usage Rate (%)'], labels=df['AI Technology'], autopct='%1.1f%%', colors=sns.color_palette('coolwarm'))
plt.title('AI Technology Usage Share in 2024')
plt.tight_layout()
plt.show()
```
